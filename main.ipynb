```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv',usecols=['movie_id','title','cast','crew'])
movies.head()
```

```python
movies=movies.merge(credits,on='title')
movies.shape
```

# Data Preprocessing

```python
movies.info()
```

```python
cols=['genres','id','keywords','title','overview','cast','crew','popularity','release_date','revenue']
movies=movies[cols]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies=movies.dropna()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast


def convert(obj):
    l=[]
    for i in ast.literal_eval(obj): #convert string list to list
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=5:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies['cast'][3]
```

```python
movies['crew'][3]
```

```python
#Director name in from crew
import ast

def fetch_dir(obj):
    l = []
    try:
        crew_list = ast.literal_eval(obj)  # Safely evaluate obj
        for i in crew_list:
            if i.get('job') == 'Director':
                l.append(i.get('name'))
                break
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing obj: {e}")
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
df=movies[['id','title','tags','popularity','release_date']]
df
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

# Text Vectorisation

```python
#Bag of words

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
print(cv.get_feature_names_out())
```

```python
#stemmming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem("eat eating eats")
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0] # 0th movie similarity with all other movies
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    index=df[df['title']==movie].index[0]
    distances=similarity[index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    
    for i in movie_list[1:6]:
        print(df.iloc[i[0]].title)
        # print(i[0])
        
```

```python
recommend('Shutter Island')
```
